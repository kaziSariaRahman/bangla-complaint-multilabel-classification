```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import seaborn as sns
matplotlib.rcParams['font.family'] = 'DejaVu Sans'

LABELS = ['Billing', 'Refund', 'Delivery', 'Customer_Service',
          'Product_Quality', 'Account', 'Technical', 'Network']
# Figure 1 — Overall F1 Comparison
metrics = ['F1 Micro', 'F1 Macro', 'F1 Weighted']
bb_overall   = [0.7735, 0.7703, 0.7729]
xlmr_overall = [0.7347, 0.7421, 0.7376]

x = np.arange(len(metrics))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(x - width/2, bb_overall,   width, label='BanglaBERT',  color='steelblue')
bars2 = ax.bar(x + width/2, xlmr_overall, width, label='XLM-RoBERTa', color='coral')
ax.set_title('BanglaBERT vs XLM-RoBERTa — Overall F1 Comparison', fontsize=13)
ax.set_ylabel('Score')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.set_ylim(0.60, 0.85)
ax.legend()
ax.bar_label(bars1, fmt='%.4f', padding=3)
ax.bar_label(bars2, fmt='%.4f', padding=3)
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('figure1_overall_f1.png', dpi=150)
plt.show()
print('figure1_overall_f1.png saved')

# Figure 2 — Per-Label F1 Comparison

bb_f1   = [0.75, 0.85, 0.80, 0.76, 0.78, 0.83, 0.66, 0.74]
xlmr_f1 = [0.61, 0.85, 0.78, 0.67, 0.78, 0.81, 0.58, 0.85]

x = np.arange(len(LABELS))
fig, ax = plt.subplots(figsize=(12, 5))
bars1 = ax.bar(x - width/2, bb_f1,   width, label='BanglaBERT',  color='steelblue')
bars2 = ax.bar(x + width/2, xlmr_f1, width, label='XLM-RoBERTa', color='coral')
ax.set_title('Per-Label F1 Score — BanglaBERT vs XLM-RoBERTa', fontsize=13)
ax.set_ylabel('F1 Score')
ax.set_xticks(x)
ax.set_xticklabels(LABELS, rotation=20, ha='right')
ax.set_ylim(0, 1.05)
ax.legend()
ax.bar_label(bars1, fmt='%.2f', padding=2, fontsize=8)
ax.bar_label(bars2, fmt='%.2f', padding=2, fontsize=8)
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('figure2_perlabel_f1.png', dpi=150)
plt.show()
print('figure2_perlabel_f1.png saved')
# Figure 3 — Hamming Loss Comparison
models = ['BanglaBERT', 'XLM-RoBERTa']
hamming = [0.0656, 0.0792]

fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(models, hamming, color=['steelblue', 'coral'], width=0.4)
ax.set_title('Hamming Loss Comparison', fontsize=13)
ax.set_ylabel('Hamming Loss (Lower is Better)')
ax.set_ylim(0, 0.12)
ax.bar_label(bars, fmt='%.4f', padding=3)
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('figure3_hamming_loss.png', dpi=150)
plt.show()
print('figure3_hamming_loss.png saved')
# Figure 4 — BanglaBERT Confusion Matrix Heatmap
tp = [93,  77,  94, 110,  76, 141,  71,  50]
fp = [36,  14,  25,  54,  19,  39,  35,   8]
fn = [26,  14,  23,  16,  24,  20,  37,  27]
tn = [640, 690, 653, 615, 676, 595, 652, 710]

cm_data = np.array([[tp[i], fp[i], fn[i], tn[i]] for i in range(8)])

fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(cm_data, annot=True, fmt='d', cmap='Blues',
            xticklabels=['TP', 'FP', 'FN', 'TN'],
            yticklabels=LABELS)
ax.set_title('BanglaBERT — Per-Label Confusion Matrix', fontsize=13)
ax.set_xlabel('Metric')
ax.set_ylabel('Label')
plt.tight_layout()
plt.savefig('figure4_confusion_matrix.png', dpi=150)
plt.show()
print('✅ figure4_confusion_matrix.png saved')
# Figure 5 — Label Distribution
label_counts = [554, 443, 529, 631, 660, 390, 671, 461]

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(LABELS, label_counts, color='steelblue')
ax.set_title('Label Distribution in Training Set', fontsize=13)
ax.set_ylabel('Count')
ax.set_xticklabels(LABELS, rotation=20, ha='right')
ax.bar_label(bars, padding=3)
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('figure5_label_distribution.png', dpi=150)
plt.show()
print('figure5_label_distribution.png saved')
```
